```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
from sklearn.neighbors import NearestNeighbors

# 위도와 경도 정보 추출
school_locations = df1[['위도', '경도']].values
company_locations = df1[['위도', '경도']].values

# 근접한 중소기업의 수를 계산하기 위한 함수
def count_nearby_companies(school_location, company_locations, radius_km=100):
    count = 0
    for company_location in company_locations:
        if geodesic(school_location, company_location).km <= radius_km:
            count += 1
    return count

# 각 폐교에 대해 근처에 있는 중소기업 수 계산
df1['근처 중소기업 수'] = df1.apply(
    lambda row: count_nearby_companies((row['위도'], row['경도']), company_locations), axis=1)

# 근접한 중소기업 수 기준으로 폐교 정렬
sorted_schools = df1.sort_values(by='근처 중소기업 수', ascending=False)

# 결과 출력
sorted_schools[['폐교명', '근처 중소기업 수']].head(20)
```

```python
import pandas as pd
from geopy.distance import geodesic
from sklearn.preprocessing import MinMaxScaler


DISTANCE_THRESHOLD1 = 100  # km
DISTANCE_THRESHOLD2 = 20  # km
DISTANCE_THRESHOLD3 = 10  # km

# 거리 계산 함수
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).kilometers

# 근접한 중소기업 수 계산 함수
def count_nearby_companies(lat, lon):
    return sum(calculate_distance(lat, lon, row['위도'], row['경도']) <= DISTANCE_THRESHOLD1 for index, row in fd1.iterrows())
# 근접한 명소 수 계산 함수
def count_nearby_attractions(lat, lon):
    return sum(calculate_distance(lat, lon, row['위도'], row['경도']) <= DISTANCE_THRESHOLD2 for index, row in travel.iterrows())
# 근접한 숙박업소 수 계산 함수
def count_nearby_accomodation(lat, lon):
    return sum(calculate_distance(lat, lon, row['위도'], row['경도']) <= DISTANCE_THRESHOLD3 for index, row in travel2.iterrows())

df1['근접_중소기업_수'] = df1.apply(lambda row: count_nearby_companies(row['위도'], row['경도']), axis=1)
df1['근접_명소_수'] = df1.apply(lambda row: count_nearby_attractions(row['위도'], row['경도']), axis=1)
df1['근접_숙박업소_수'] = df1.apply(lambda row: count_nearby_accomodation(row['위도'], row['경도']), axis=1)

# 필요한 열만 선택하여 데이터프레임 생성
df_analysis = df1[['근접_중소기업_수', '근접_명소_수', '근접_숙박업소_수', '건물연면적', '대지']]

# 데이터 정규화 (0-1 스케일)
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_analysis), columns=df_analysis.columns)

# 가중치 설정
weights = {
    '근접_중소기업_수': 0.35,
    '근접_명소_수': 0.3,
    '근접_숙박업소_수':0.2,
    '건물연면적': 0.1,
    '대지': 0.05
}

# 총점 계산
df_normalized['총점'] = (df_normalized['근접_중소기업_수'] * weights['근접_중소기업_수'] +
                         df_normalized['근접_명소_수'] * weights['근접_명소_수'] +
                         df_normalized['근접_숙박업소_수'] * weights['근접_숙박업소_수'] +
                         df_normalized['건물연면적'] * weights['건물연면적'] +
                         df_normalized['대지'] * weights['대지'])

# 상위 10개 폐교 선택
df1['총점'] = df_normalized['총점']
top_schools = df1.nlargest(10, '총점')

top_schools
```
